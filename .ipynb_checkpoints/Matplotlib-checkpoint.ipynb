# Matplotlib

matplotlib.org

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
#if not in notebook, add plt.show() at the end
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

## Functional method of creating plots

```python
plt.plot(x, y, 'g--')
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.title('Title')
plt.show()
```

### Creating subplots

```python
plt.subplot(1,2,1) #colums, rows, subplot
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

## Object Oriented API method

```python
#create figures and call methods off of them
fig = plt.figure() #this creates a figure object - a blank canvas
```

```python
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 
#now we add a set of axes to our canvas
#by calling the add_axes method of our fig-object
#arguments specify 'the percent' of the canvas you want to take
fig
```

```python
axes.plot(x,y)
#on the axes, we can plot our data
fig
```

```python
#So all together: 
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 
#left, bottom, with, heigth (all in range from 0 to 1)
axes.plot(x,y)

axes.set_xlabel('x Label')
axes.set_ylabel('y Label')
axes.set_title('Title')
```

```python
#Now multiple axes :D 
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) 

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 2) #subplot is an axes-manager
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
axes #axes is an array of matplotlib-axes
#therefore, we can iterate through it!
#and we can index it with bracket notation axes[0].some_method
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
#iterating over the axes array
for i in axes: 
    i.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
#indexing
axes[0].plot(x,y)
axes[1].plot(y,x)
#in the same way, we can add titles and so on
```

## Figure Size and DPI

DPI: Dots/Pixels per inch

```python
fig = plt.figure(figsize=(7,2), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(7,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
fig.tight_layout()
```

### Saving plots

```python
fig.savefig('my_plot.png',dpi=200)
```

## Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='x squared')
ax.plot(x, x**3, label='x cubed')

ax.legend(loc=(0.2,0.3)) #sets lower left corner to this location
```

## Customize the look of the plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange', linewidth=13, alpha=0.3, linestyle='-.',
        marker='*', markersize=30, markerfacecolor='gold', 
       markeredgewidth=2, markeredgecolor='orange') 
#color takes RGB hex codes as well
#alpha is transparency
#linewidth shorthand lw=2 works as well
#linestyles (short ls): -, , -., :, steps
#marker: '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
#markersize
#markerfacecolor
#markeredgewith
#markeredgecolor

# custom dash
line, = ax.plot(x, x**2+3, color="black", lw=1.50)
line.set_dashes([10, 15, 20, 15]) # format: line length, space length, ...
```

## Plot range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x=np.linspace(-4, 4, 100)
ax.plot(x,np.sinc(x))
ax.set_xlim((-2, 3))
ax.set_ylim((0,1))
```
